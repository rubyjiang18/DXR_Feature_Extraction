```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import os
from os.path import isfile, join
from PIL import Image
import os.path, sys
```

```python
def Get_each_frame_contours(frame,ExpNum):
    frame = str(frame)
    ExpNum = str(ExpNum)
    #img = cv.imread('/Users/rubyjiang/Desktop/39_bg/' + 'frame' + frame + '.jpg')
    img = cv.imread('/Users/rubyjiang/Desktop/' + ExpNum + '_bg/' + 'frame' + frame + '.jpg')
    grayscaled = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    #
    #The second argument is the threshold value which is used to classify the pixel values.
    retval, thresholdG = cv.threshold(grayscaled, 5, 255, cv.THRESH_BINARY) 
    # fastNlMeansDenoising
    # h : parameter deciding filter strength. Higher h value removes noise better, but removes details of image also. (10 is ok)
    # adjust two h to connect keyholes,	Parameter regulating filter strength. Big h value perfectly removes noise but also removes image details, smaller h value preserves details but also preserves some noise
    # do not change 7,21
    dst = cv.fastNlMeansDenoising(thresholdG,None,70,7,21) 
    dst2 = cv.fastNlMeansDenoising(dst,None,100,7,21)
    retval, thresh2 = cv.threshold(dst2, 150, 255, cv.THRESH_BINARY)
    #Find contour
    contours,hierarchy = cv.findContours(thresh2, 1, 2)
    plt.imshow(thresh2, cmap='Greys_r')
    return contours
```

```python
#To find the different features of contours, like area, perimeter, centroid, bounding box etc
def Get_Cnt_Features(cnt):
    #
    M = cv.moments(cnt)
    #
    microns = 1.932
    #
    if M['m00']==0:
        cx = 0
        cy = 0
    else: 
        cx = int(M['m10']/M['m00'])
        cy = int(M['m01']/M['m00'])
    area = cv.contourArea(cnt)
    area = area*microns
    perimeter = cv.arcLength(cnt,True)
    perimeter = perimeter*microns
    (x,y),radius = cv.minEnclosingCircle(cnt)
    center = (int(x),int(y))
    radius = int(radius)
    radius = radius*microns
    x,y,w,h = cv.boundingRect(cnt)
    aspect_ratio = float(w)/h
    area = cv.contourArea(cnt)
    x,y,w,h = cv.boundingRect(cnt)
    rect_area = w*h
    extent = float(area)/rect_area
    area = cv.contourArea(cnt)
    hull = cv.convexHull(cnt)
    hull_area = cv.contourArea(hull)
    if hull_area == 0:
        solidity = 0
    else:
        solidity = float(area)/hull_area
    area = cv.contourArea(cnt)
    equi_diameter = np.sqrt(4*area/np.pi)
    equi_diameter = equi_diameter*microns
    topmost2 = (cnt[cnt[:,:,1].argmin()][0])
    bottommost2 = (cnt[cnt[:,:,1].argmax()][0])
    depth = bottommost2-topmost2
    depth = depth*microns
    d = [(depth)]
    dx = [x for x, y in d]
    dy = [y for x, y in d] #keyhole depth
    rightmost2 = (cnt[cnt[:,:,0].argmax()][0])
    leftmost2 = (cnt[cnt[:,:,0].argmin()][0])
    width = rightmost2 - leftmost2
    width = width*microns
    w = [(width)]
    wx = [x for x, y in w] #keyhole width
    wy = [y for x, y in w] 
    
#     print('frame is:',frame)
#     print('cx,cy is:',cx,cy)
#     print('area is:',area)
#     print('perimeter is:',perimeter)
#     print('radius is:',radius)
#     print('aspect_ratio is:',aspect_ratio)
#     print('extent:',extent)
#     print('solidity is:',solidity)
#     print('equi_diameter:',equi_diameter)
#     print('key hole depth and width are:',dy,wx,'respectively')
    return cx,cy,area,perimeter,radius,aspect_ratio,extent,solidity,equi_diameter,dy[0],wx[0]
    #return area,perimeter,radius,aspect_ratio,extent,solidity,equi_diameter,dy[0],wx[0]
```

### Get_Keyhole_Dimensions for keyholes w/ ot w/o pores.

```python
#Get the keyhole object out of pores for slender keyholes, for chubby keyholes, the only object is the keyhole.
def Get_each_frame_Keyhole_Dimensions(contours):
    #make lists to hold values for each object
    cx_list = []
    cy_list = []
    area_list = []
    perimeter_list = []
    radius_list = []
    aspect_ratio_list = []
    extent_list = []
    solidity_list = []
    equi_diameter_list = []
    keyhole_depth_list = []
    keyhole_width_list = []
    #loop through every objects and append values
    i = 0
    #for cnt in contours:
    for cnt in contours:
        # skip the small cnt
        if len(cnt) < 5:
            continue
        if cv.contourArea(cnt) <= 100:
            continue
        
        print('#################### Object ', i, '####################')
        cx,cy,area,perimeter,radius,aspect_ratio,extent,solidity,equi_diameter,dy,wx = Get_Cnt_Features(cnt)
        # Append these features to their list
        cx_list.append(cx)
        cy_list.append(cy)
        area_list.append(area)
        perimeter_list.append(perimeter)
        radius_list.append(radius)
        aspect_ratio_list.append(aspect_ratio)
        extent_list.append(extent)
        solidity_list.append(solidity)
        equi_diameter_list.append(equi_diameter)
        keyhole_depth_list.append(dy)
        keyhole_width_list.append(wx)
        i +=1
    # turn lists into numpy array
    cxs = np.asarray(cx_list)
    cys = np.asarray(cy_list)
    areas = np.asarray(area_list)
    perimeters = np.asarray(perimeter_list)
    radiuss = np.asarray(radius_list)
    aspect_ratios = np.asarray(aspect_ratio_list)
    extents = np.asarray(aspect_ratio_list)
    soliditys = np.asarray(solidity_list)
    equi_diameters = np.asarray(equi_diameter_list)
    keyhole_depths = np.asarray(keyhole_depth_list)
    keyhole_widths = np.asarray(keyhole_width_list)
    #Find the keyhole object with the the maximum area.
    index = np.argmax(areas)
    print('Index of the object with the largest area is:', index)
    #index = index-1
    #Fill the list with keyhole dimensions
    keyhole_dimensions = [
                            cxs[index],cys[index],
                            areas[index],
                          perimeters[index],radiuss[index],
                          aspect_ratios[index],extents[index],
                            soliditys[index],equi_diameters[index],
                            keyhole_depths[index], keyhole_widths[index]
                         ]
    keyhole_dimensions = np.asarray(keyhole_dimensions)
    return keyhole_dimensions 
```

### Get all frames keyhole dimensions

```python
def FinalGet(ExpNum):
    frames = list(range(160,359))#513 frames in total
    All_frames_keyhole_Dimensions = []
    for frame in frames:
        contours = Get_each_frame_contours(frame,ExpNum)
        keyhole_dimensions = Get_each_frame_Keyhole_Dimensions(contours)
        All_frames_keyhole_Dimensions.append(keyhole_dimensions)
    All_frames_keyhole_Dimensions = np.asarray(All_frames_keyhole_Dimensions)
    return All_frames_keyhole_Dimensions
```
